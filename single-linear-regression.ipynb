```python
# Imports
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import scipy.stats
pd.options.mode.chained_assignment = None

# Get Weight and MPG.highway from car93
cars = pd.read_csv('data/Cars93.csv')[['Weight', 'MPG.highway']]

# Get correlation between fields
cars.corr()
```

```python
# Create graphic comparing weight and MPG highway
fig1, ax1 = plt.subplots()
ax1.scatter(cars['Weight'], cars['MPG.highway'])
ax1.set_xlabel('Vehicle Weight')
ax1.set_ylabel('MPG (Highway)')
ax1.set_title('Vehicle Weight VS MPG on Highway')
```

```python
# Calculate linear regression line
x = cars['Weight'].to_numpy()
y = cars['MPG.highway'].to_numpy()
slope, intercept, r, p, stderr = scipy.stats.linregress(x,y)
line = f'Slope: {slope:.2f}\nIntercept: {intercept:.2f}\nCorrelation coefficient: r={r:.2f}'
print(line)
```

```python
# Create graphic comparing weight and MPG highway
fig2, ax2 = plt.subplots()
ax2.scatter(cars['Weight'], cars['MPG.highway'])
ax2.set_xlabel('Vehicle Weight')
ax2.set_ylabel('MPG (Highway)')
ax2.set_title('Vehicle Weight VS MPG on Highway')
ax2.plot(x, intercept + slope * x, label=line, color='r')
ax2.legend(facecolor='white')
plt.show()
```
