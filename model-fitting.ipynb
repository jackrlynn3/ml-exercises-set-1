```python
# Imports
import pandas as pd
from sklearn.model_selection import train_test_split
pd.options.mode.chained_assignment = None

# Get data
data = pd.read_csv('data/song_data_clean.csv')

# Training toward song length
X = data.copy().drop(columns=['Length (Duration)'])
y = data['Length (Duration)'].copy()

# Get a 30%/70% train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```
